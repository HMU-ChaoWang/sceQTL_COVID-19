```r
##-----------------------------1.celltype eqtl文件读入，转化为ensembleID----------------------------------------------##    
    celltype_eqtl<-read.table(allfiles_filepath[3],header=T,sep="\t") ##文件读入
    eGene_num<-length(unique(celltype_eqtl$gene)) ##eGene数目
    eSNP_num<-length(unique(celltype_eqtl$snps)) ##eSNP数目
    eqtl_num<-dim(celltype_eqtl)[1] ##eqtl数目
    intersect_genesymbol<-emsemblelist_Rpackages[which(emsemblelist_Rpackages$SYMBOL%in%unique(celltype_eqtl$gene)),] ##交集smybol的ensemble信息
    celltype_rsID_gene<-as.data.frame(celltype_eqtl[,1:2]) ##celltype eqtl前两列
    
    
        ##symbol转化为ensemble
    for(j in 1:dim(intersect_genesymbol)[1]){
         loc1<-which(celltype_rsID_gene[,2]%in%intersect_genesymbol[j,3])
    
         celltype_rsID_gene[loc1,2]<-intersect_genesymbol[j,1]
    }
```

```r
##---------------------------2.celltype eqtl与DICE eqtl的交集---------------------------------------------------------##    

    tissue<-"blood"  ##组织名称
    celltype<-strsplit(strsplit(allfiles[i],split="_cis_eqtls")[[1]][1],split=paste0("_",tissue,".txt"))[[1]][1]  ##细胞类型名称
    DICE_eqtl<-CD8
    
    
    
    GETxv8_paste_rsIDensemble<-paste0(DICE_eqtl$first_column,":",DICE_eqtl$second_column)##GETx前两列转化成 rsID:ensemble
    celltype_paste_rsIDensemble<-paste0(celltype_rsID_gene$snps,":",celltype_rsID_gene$gene) ##celltype eqtl 前两列转化成 rsID:ensemble
    
    intersect_eqtl_num<-length(intersect(GETxv8_paste_rsIDensemble,celltype_paste_rsIDensemble)) ##交集eqtl个数
    inter_pt<-intersect_eqtl_num/eqtl_num ##交集eqtl 占比
    
```

```r
intersect_eqtl_num
inter_pt
eqtl_num
```

```r
##---------------------------3.celltype eqtl与GETx eqtl的交集信息-----------------------------------------------------##    
    intersect_eqtl<-intersect(GETxv8_paste_rsIDensemble,celltype_paste_rsIDensemble) ##交集eqtl
    rsIDensemble_celltype<-lapply(intersect_eqtl,intersect_eqtl_split) ##分割交集
    rsID_ensembleID<-c()
    ##分割后形成两列
    for(k in 1:length(rsIDensemble_celltype)){
       rsID_ensembleID<-rbind(rsID_ensembleID,rsIDensemble_celltype[[k]])
    }
    colnames(rsID_ensembleID)<-c("rsID","EnsembleID") 
    rsID_ensembleID<-as.data.frame(rsID_ensembleID)
    
    
    intersect_eGene <-unique(rsID_ensembleID$EnsembleID)
    
    intersect_eGene_num<-length(unique(rsID_ensembleID$EnsembleID)) ##交集eGene数目
    intersect_eSNP_num<-length(unique(rsID_ensembleID$rsID)) ##交集eGene数目

    
    rsID_GeneID<-rsID_ensembleID
    
    for(x in 1:length(intersect_eGene)){
        ensemble_loc<-which(emsemblelist_Rpackages$ENSEMBL%in%intersect_eGene[x])
        gene<-emsemblelist_Rpackages[ensemble_loc[1],3]
        ensemble_loc2<-which(rsID_GeneID$EnsembleID%in%intersect_eGene[x])
        rsID_GeneID[ensemble_loc2,2]<-gene
    }
   colnames(rsID_GeneID)<-c("rsID","Symbol")
    
    
    GTExv8_result<-c()
    for(y in 1:length(intersect_eqtl)){
    GTExv8_loc<-which(GETxv8_paste_rsIDensemble%in%intersect_eqtl[y])
    GTExv8_result<-c(GTExv8_result,GTExv8_loc)
    }
    ##GETx eqtl的交集信息
    intersect_GTExv8<-DICE_eqtl[GTExv8_result,]
    intersect_GTExv8_rsIDSymbol<-cbind(rsID_GeneID,intersect_GTExv8[,c(1,2,3,4,7)])
    
   write.table(intersect_GTExv8_rsIDSymbol,paste0(result_path,celltype,"_cis_",tissue,"_DICE_CD8.txt"),quote=F,sep="\t",row.names=F)
        
    celltype_result<-c()
    for(z in 1:length(intersect_eqtl)){
    celltype_loc<-which(celltype_paste_rsIDensemble%in%intersect_eqtl[z])
    celltype_result<-c(celltype_result,celltype_loc)
    }
    ##celltype eqtl的交集信息
    intersect_celltype<-celltype_eqtl[celltype_result,]
    intersect_celltype_rsIDSymbol<-intersect_celltype
    
    write.table(intersect_celltype_rsIDSymbol,paste0(result_path,celltype,"_Tissue_",tissue,"_sc_CD8.txt"),quote=F,sep="\t",row.names=F)
```

```r
result_path
```

```r
##-------------------------------4.判断交集eqtl调控方向是否一致-----------------------------------------------------##    
    
    GTEx_sign<-sign(intersect_GTExv8_rsIDSymbol$beta_meta)
    celltype_sign<-sign(intersect_celltype_rsIDSymbol$beta)
    
    consistent_num<-0
    for(a in 1:length(GTEx_sign)){
        if(GTEx_sign[a]==celltype_sign[a]){
        consistent_num<-consistent_num+1
        }    
    }
    
    con_pt<-consistent_num/intersect_eqtl_num ##调控方向一致比例  
    
    result_celltype<-cbind(celltype,tissue,eqtl_num,eGene_num,eSNP_num,intersect_eqtl_num,intersect_eGene_num,
                         intersect_eSNP_num,inter_pt,consistent_num,con_pt)
    
    result_allcelltype_all<-rbind(result_allcelltype_all,result_celltype)
```
