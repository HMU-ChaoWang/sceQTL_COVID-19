# 定义公式，求取b_meta,se_meta,z_meta

```r
calculate_formula <- function(beta, se) {
  numerator <- 0
  denominator <- 0
  
  for (k in 1:length(beta)) {
    numerator <- numerator + beta[k] / (se[k] ^ 2)
    denominator <- denominator + 1 / (se[k] ^ 2)
  }
  
  result <- numerator / denominator
  return(result)
}

calculate_formula1 <- function(se) {
  denominator <- 0
  
  for (k in 1:length(se)) {
    denominator <- denominator + 1 / (se[k] ^ 2)
  }
  
  result <- 1 / denominator
  return(result)
}

task1 <- function(beta_values,se_values) {
  
  # 调用函数计算结果
  beta_result <- calculate_formula(beta_values, se_values)
  
  # 调用函数计算结果
  se_result <- calculate_formula1(se_values)
  
  Z_result <- (beta_result / se_result)
  res<-c(beta_result,se_result,Z_result )
   
    
  #print(paste("beta_result =", beta_result))
#print(paste("se_result =", se_result))
#print(paste("Z_result =", Z_result))
    
}
```
