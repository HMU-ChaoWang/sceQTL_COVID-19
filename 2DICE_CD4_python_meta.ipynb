```python
a = pd.DataFrame(df1['num_new'].value_counts()).reset_index()
a.columns = ['num_new', 'count']
a
```

```python
b = pd.merge(df1, a, left_on='num_new', right_on='num_new')
b
```

```python
z = b['id_gene'].unique()
z
```

```python
def calculate_formula(beta, se):
    numerator = 0
    denominator = 0
    
    for k in range(len(beta)):
        numerator += beta[k] / (se[k] ** 2)
        denominator += 1 / (se[k] ** 2)
    
    result = numerator / denominator
    return result


def calculate_formula1(se):
    denominator = 0
    
    for k in range(len(se)):
        denominator += 1 / (se[k] ** 2)
    
    result = 1 / denominator
    return result


def task1(beta_values, se_values):
    # 调用函数计算结果
    beta_result = calculate_formula(beta_values, se_values)
  
    # 调用函数计算结果
    se_result = calculate_formula1(se_values)
  
    Z_result = beta_result / se_result
    res = [beta_result, se_result, Z_result]
    return res
```

```python
cccCD4 = []
sssCD4 = []

for i in range(len(z)):
    loc = b[b["id_gene"].isin([z[i]])]
    B = loc["Beta"].values
    SE = loc["se"].values
    result = np.asarray(task1(B, SE))
    zzzCD4 = z[i]
    sssCD4 = [zzzCD4] + result.tolist()
    cccCD4.append(sssCD4)

cccCD4 = pd.DataFrame(cccCD4)
```
