```python
import pandas as pd
import numpy as np
```

```python
GTEx_blood= pd.read_csv('/data_alluser/CXY/keti/singlecell_eqtl/eqtl/result/GTEx/Whole_Blood.signifpairs.txt',sep="\t")
GTEx_blood
```

```python
GTExv8 = pd.read_csv('/data_alluser/CXY/keti/singlecell_eqtl/eqtl/result/GTEx/PBMCv8_rsIDchrposensemble.txt',sep="\t")
GTExv8
```

```python
eQTL = pd.read_csv('/data_alluser/CXY/keti/singlecell_eqtl/eqtl/result/GTEx/eQTL_snpsloc_new2.txt',sep="\t")
eQTL
```

```python
#根据rsID这列把两个dataFrame合并，以左侧的dataFrame为主
left_intersected_df = pd.merge(GTExv8, eQTL, on='rsID', how='left')
left_intersected_df
```

```python
#把snp这列为空的值用rsID这列的值替代，不为空的值保留不变
left_intersected_df['snp'] = np.where(pd.isnull(left_intersected_df['snp']), left_intersected_df['rsID'], left_intersected_df['snp'])
left_intersected_df
```

```python
GTExv8['rsID'] = left_intersected_df['snp']
GTExv8
```

```python
GTExv8.to_csv("/data_alluser/CXY/keti/singlecell_eqtl/eqtl/result/GTEx/PBMCv8_ensemblechrpossnp.txt", index=False, sep="\t", header=True)
```
